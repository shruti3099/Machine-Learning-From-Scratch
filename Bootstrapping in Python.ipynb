## What is bootstrapping?

Bootstrapping is a sampling method where a population parameter for a large dataset is determined by calculating the parameter for small samples drawn from the dataset, with replacement.

Sampling with replacement means a data point in a drawn sample can reappear in future drawn samples as well. 

Bootstrapping is used in places where we have a large dataset and it is tedious to compute the population parameters for all the data samples. When we use small smaples instead, the computation time is hugely reduced, and the results are very close to what you would get if we took the entire dataset into account for calculation.

For example, let's say we have 1000 students in a school and we want to estimate their mean height.

## How is bootstrapping implemented?

The entire process of bootstrapping can be broken down into 5 steps:

1. Choose the dataset to perform bootstrapping upon.
2. Define the size of bootstrap sample and number of iterations to be carried out.
3. For each iteration, draw random data points from dataset (bootstrap sample) with replacement.
4. Calculate the required parameter for each sample.
5. Take the average of all samples' parameters to find the population parameter for entire dataset.

Now that you are clear with the theory of bootstrapping, let's look at the pseudocode for these steps.

`
sample_statistics = []`

`for i in bootstraps: #bootstraps=number of iterations
	sample = select_sample_with_replacement(dataset)
	stat = calculate_statistic(sample)
	sample_statistics.append(stat)`

`population_statistic = mean(sample_statistics)`

## Worked Example

For implementing the pseudocode we will generate a dataset of 1000 random numbers that follow a normal disribution, using `np.random.normal()`. 
We will calculate the mean for this dataset using `np.mean()` function as well estimate the mean though bootstrapping.

```python
import numpy as np
import random
# normal distribution 
data = np.random.normal(loc= 100, scale=1, size=10000)

np.mean(data)
```

As obtained by `np.mean()`, the mean of our dataset is 100.01. Let's try to estimate the same using bootstrap sampling.
We are taking a sample size of 5 and performing 40 iterations.

```python
sample_means = []

# Bootstrap Sampling
for i in range(40):
    sample = random.sample(data.tolist(), 5)
    mean = np.mean(sample)
    
    sample_means.append(mean)

np.mean(sample_means)
```

The estimated mean of dataset by using bootstrapping is 100.08! As you can see, it is very close to the actual mean which means bootstrapping is an effective way to save computing time on large datasets.
